```python
import pandas as pd
```

```python
dados = pd.read_json('/content/drive/MyDrive/ALURA/dados_hospedagem.json')
dados.head()
```

```python
dados = pd.json_normalize(dados['info_moveis'])
dados
```

```python
colunas = list(dados.columns)
colunas
```

```python
dados = dados.explode(colunas[3:])
dados
```

```python
dados.reset_index(inplace=True, drop=True)
dados.head()
```

```python
dados.info()
```

```python
import numpy as np
```

```python
# Transformando os dados das colunas para int64
col_numericas = ['max_hospedes', 'quantidade_banheiros', 'quantidade_quartos', 'quantidade_camas']
dados[col_numericas] = dados[col_numericas].astype(np.int64)
```

```python
dados['avaliacao_geral'] = dados['avaliacao_geral'].astype(np.float64)
```

```python
# Limpa os valores das colunas de preços e taxas, removendo os caracteres '$' e ',' e espaços em branco.
# Utiliza-se o método map para aplicar uma função lambda que faz essa limpeza célula a célula,
# preparando os dados para conversão numérica.
col_precos = ['preco', 'taxa_deposito', 'taxa_limpeza']
dados[col_precos] = dados[col_precos].map(lambda x: x.replace('$','').replace(',','').strip())
```

```python
# Transformando os dados das colunas para float64
dados[col_precos] = dados[col_precos].astype(np.float64)
```

```python
dados.info()
```

```python
# Utilizando método str para transformação das letras para minúsculas
dados['descricao_local'] = dados['descricao_local'].str.lower()
```

```python
# Substitui por espaço todos os caracteres que não sejam letras (a-z, A-Z), números (0-9), hífen (-) ou apóstrofo (')
dados['descricao_local'] = dados['descricao_local'].str.replace('[^a-zA-Z0-9\-\']',' ', regex=True)
```

```python
# Substitui por espaço o hífen "-" que não possui caracteres alfanuméricos antes ou depois, usando regex
dados['descricao_local'] = dados['descricao_local'].str.replace('(?<!\w)-(?!\w)',' ', regex=True)
```

```python
# Tokenização das strings na coluna "descricao_local"
dados['descricao_local'] = dados['descricao_local'].str.split()
```

```python
# Remove os caracteres {, } e " da coluna 'comodidades' usando regex
dados['comodidades'] = dados['comodidades'].str.replace('\{|}|\"','', regex=True)
```

```python
# Tokenização das strings na coluna "comodidades"
dados['comodidades'] = dados['comodidades'].str.split(',')
```

```python
dados['descricao_vizinhanca'] = dados['descricao_vizinhanca'].str.lower()
```

```python
dados['descricao_vizinhanca'] = dados['descricao_vizinhanca'].str.replace('[^a-zA-Z0-9\-\']', ' ', regex=True)
```

```python
dados['descricao_vizinhanca'] = dados['descricao_vizinhanca'].str.replace('(?<!\w)-(?!\w)', ' ', regex=True)
```

```python
# Remove aspas simples isoladas no texto
dados['descricao_vizinhanca'] = dados['descricao_vizinhanca'].str.replace(r"\b'\b", '', regex=True)
```

```python
import re
import nltk
from nltk.corpus import stopwords

# Obtém a lista de stopwords em inglês
nltk.download('stopwords')
stopwords_en = stopwords.words('english')

# Cria padrão regex para detectar palavras inteiras que são stopwords
padrao = r'\b(?:' + '|'.join(stopwords_en) + r')\b'

# Substitui todas as stopwords por espaço em branco na coluna, usando regex
dados['descricao_vizinhanca'] = dados['descricao_vizinhanca'].str.replace(padrao, ' ', regex=True)
```

```python
# Tokenização das strings na coluna "descricao_vizinhanca"
dados['descricao_vizinhanca'] = dados['descricao_vizinhanca'].str.split()
```
